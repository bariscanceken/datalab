## ***EDA***

```python
#Top Rating Product Name
dfc_top_rating = dfc[["rating", "product_name"]].sort_values(by = "rating", ascending = False).head(10)
dfc_top_rating
```

```python
# Top Discount Amount Product Name
dfc_top_price = dfc[["rating", "rating_count", "actual_price", "discounted_price", "discount_amount", "product_name"]].sort_values(by = "discount_amount", ascending = False).head(10)
dfc_top_price_plt = dfc[["discounted_price", "discount_amount","product_name"]].sort_values(by = "discount_amount", ascending = False).head(10)

dfc_top_price
```

```python
names = dfc_top_price_plt["product_name"].str[:10] + "..."
custom_labels_blue = [a for a in dfc_top_price["rating"]]
ax = dfc_top_price_plt.set_index(names).plot(kind='bar', stacked=True, color=[cyberpunk_palette[0], cyberpunk_palette[2]])
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1], labels=custom_labels_blue, label_type='center')
ax
```

```python
# Sales Differences Between Categories
dfc_top_sales_category = dfc.groupby(["category"])["category_id"].sum().sort_values(ascending=False)
dfc_top_sales_category.reset_index()
```

```python
df_sns = pd.DataFrame(dfc_top_sales_category)
df_sns
```

```python
ax = sns.scatterplot(data=df_sns)
ax.set(xlabel='Categories')
ax.set(ylabel='Sales')

ax.set_xticklabels([])
ax
```
